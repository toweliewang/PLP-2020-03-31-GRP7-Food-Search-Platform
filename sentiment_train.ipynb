```
def make_dir(dirName):

  if not os.path.exists(dirName):
      os.mkdir(dirName)
      print("Directory " , dirName ,  " Created ")
  else:    
      print("Directory " , dirName ,  " already exists")

make_dir('train')
make_dir('test')
make_dir('model_lstm')
```

```
yelp_google = pd.read_csv('yelp_label.csv',header=0,nrows=19000,encoding="utf-8")
yelp_labelled = pd.read_csv('./yelp_labelled.txt',sep='\t', header=None, names=['text','label'])
df = pd.concat([yelp_google,yelp_labelled]).sample(frac=1,random_state=42)
df['label'].value_counts()
```

```

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.20, random_state=42)
pd.concat([X_train,y_train],axis=1).to_csv('./train/train.csv')
pd.concat([X_test,y_test],axis=1).to_csv('./test/test.csv')
```
